```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def get_variance(n):
    normal_array = np.random.normal(size=(n,n))
    means = np.mean(normal_array, axis=0)
    variance = np.var(means)
    return variance

def display_data():
    n_vals = np.linspace(100, 1000, 10, dtype=np.int64)
    variances = []
    for n in n_vals:
        variances.append(get_variance(n))
    plt.plot(n_vals, variances)
    plt.show()

display_data()
```

```python
def display_trig():
    domain = np.linspace(-2 * np.pi, 2 * np.pi, 1000)
    sin = np.sin(domain)
    cos = np.cos(domain)
    arctan = np.arctan(domain)
    plt.plot(domain, sin)
    plt.plot(domain, cos)
    plt.plot(domain, arctan)
    plt.show()

display_trig()
```

```python
def display_curves():
    domain1 = np.linspace(-2, 0.999, 500)
    domain2 = np.linspace(1.001, 6, 500)
    curve1 = 1 / (domain1 - 1)
    curve2 = 1 / (domain2 - 1)
    plt.plot(domain1, curve1, 'm--', lw=4)
    plt.plot(domain2, curve2, 'm--', lw=4)
    plt.ylim(-6, 6)
    plt.show()

display_curves()
```

```python
def display_sin():
    domain = np.linspace(0, 2*np.pi, 1000)
    graph1 = np.sin(domain)
    graph2 = np.sin(2 * domain)
    graph3 = 2 * np.sin(domain)
    graph4 = 2 * np.sin(2 * domain)
    
    plt.subplot(221)
    plt.axis([0, 2*np.pi, -2, 2])
    plt.title("sin(x)")
    plt.plot(domain, graph1, 'g-')
    
    plt.subplot(222)
    plt.axis([0, 2*np.pi, -2, 2])
    plt.title("sin(2x)")
    plt.plot(domain, graph2, 'r--')
    
    plt.subplot(223)
    plt.axis([0, 2*np.pi, -2, 2])
    plt.title("2sin(x)")
    plt.plot(domain, graph3, 'b--')
    
    plt.subplot(224)
    plt.axis([0, 2*np.pi, -2, 2])
    plt.title("2sin(2x)")
    plt.plot(domain, graph4, 'm.')
    
    plt.suptitle("Various sin(x) graphs")
    
    plt.tight_layout()
    
    plt.show()

display_sin()
```

```python
def display_FARS_data():
    data = np.load('../FARS.npy')
    hours = data[:, 0]
    longitudes = data[:, 1]
    latitudes = data[:, 2]
    
    plt.subplot(121)
    plt.plot(longitudes, latitudes, "k,")
    plt.axis("equal")
    plt.xlabel('Longitudes')
    plt.ylabel('Latitudes')
    plt.show()
    plt.close()
    
    plt.subplot(122)
    plt.hist(hours, bins=24, range=[0, 24])
    plt.xlim(0, 24)
    plt.xlabel("Hours")
    plt.ylabel("Car Accidents")
    plt.show()
    plt.close()

display_FARS_data()
```

```python
def display_color():
    domain = np.linspace(-2*np.pi, 2*np.pi, 1000)
    X, Y = np.meshgrid(domain, domain)
    Z = (np.sin(X) * np.sin(Y)) / (X * Y)
    
    plt.subplot(121)
    plt.axis([-2*np.pi, 2*np.pi, -2*np.pi, 2*np.pi])
    plt.pcolormesh(X, Y, Z, cmap="Spectral")
    plt.colorbar()
    
    plt.subplot(122)
    plt.axis([-2*np.pi, 2*np.pi, -2*np.pi, 2*np.pi])
    plt.contour(X, Y, Z, 15, cmap="Spectral")
    plt.colorbar()
    
    plt.show()

display_color()
```
